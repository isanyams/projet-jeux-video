#### Nettoyage des datasets games

```python

#Ajout d'une colonne plateforme
games_playstation['plateforme'] = 'playstation'
games_steam['plateforme'] = 'steam'
games_xbox['plateforme'] = 'xbox'

#concatenation des dataset games
games = pd.concat(objs = [games_playstation, games_steam, games_xbox], axis = 0, ignore_index=True)

#remplacement des valeurs vides par unknown
games = games.fillna({'title': 'unknown', 'platform': 'unknown', 'developers' : 'unknown', 'publishers': 'unknown', 'genres': 'unknown', 'supported_languages': 'unknown'})

#Ajout d'une colonne contenant le genre principal
games['main_genre'] = games['genres'].str.strip('[]').str.replace("'", "").str.split(',').str[0]

#Conversion des colonnes publisher, developers et supported_languages en list
games['developers'] = games['developers'].str.strip('[]').str.replace("'", "").str.split(',')
games['publishers'] = games['publishers'].str.strip('[]').str.replace("'", "").str.split(',')
games['supported_languages'] = games['supported_languages'].str.strip('[]').str.replace("'", "").str.split(',')

#nombre de langue par jeu
games['supported_languages_count'] = len(games['supported_languages'])

#conversion de la date release_date en format date
games['release_date'] = pd.to_datetime(arg=games['release_date'])


#ajout d'une colonne année de sortie
games['release_year'] = games['release_date'].dt.year

#l'identifiant du jeu est suffixé par la plateforme
games['gameid'] = games['gameid'].astype('str') + '_' + games['plateforme']
```

#### Nettoyage des dataset players

```python
#Ajout d'une colonne plateforme
players_playstation['plateforme'] = 'playstation'
players_steam['plateforme'] = 'steam'
players_xbox['plateforme'] = 'xbox'

#Concaténation des dataset plateforme
players = pd.concat([players_playstation, players_steam, players_xbox], axis = 0, ignore_index= True)

#Suppression de la colonne nickname
players.drop('nickname', axis=1, inplace=True)

#Conversion de la colonne created au format date
players['created'] = pd.to_datetime(arg = players['created'],errors='coerce')

#Ajout de la colonne année de création
players['created_year'] = players['created'].dt.year

#remplacement des valeurs vide par unknown
players = players.fillna({'country': 'unknown', 'created': 'unknown', 'created_date' : 'unknown', 'created_time' : 'unknown'})

#l'identifiant playerid est suffixé par la plateforme
players['playerid'] = players['playerid'].astype('str') + '_' + players['plateforme']
```

#### Nettoyage du dataset price

```python
#Ajout d'une colonne plateforme
prices_playstation['plateforme'] = 'playstation'
prices_steam['plateforme'] = 'steam'
prices_xbox['plateforme'] = 'xbox'

#Concatenation des dataset price
prices = pd.concat([prices_playstation, prices_steam, prices_xbox], axis = 0, ignore_index= True)

#Conversion de la colonne date_acquired en date
prices['date_acquired'] = pd.to_datetime(arg = prices['date_acquired'],errors='coerce')

#Ajout de l'année du prix
prices['acquired_year'] =  prices['date_acquired'].dt.year

#L'identifiant du jeu est suffixé par la plateforme
prices['gameid'] = prices['gameid'].astype('str') + '_' + prices['plateforme']
```

```python
prices
```

#### Conversion des prix en dollar

```python
# Remplacer les valeurs manquantes par Null

prices = prices.replace(np.nan, 'Null')

# Supprimer les gameid en double dans le dataframe prices

price_without_double = prices.drop_duplicates(subset='gameid')

# Copier le DataFrame original
prices_usd = price_without_double.copy()

# Remplacer 'Null' par np.nan
prices_usd = prices_usd.replace('Null', np.nan)

# Colonnes de devise
cols = ['usd', 'eur', 'gbp', 'jpy', 'rub']

# Convertir les colonnes en float
for col in cols:
    prices_usd[col] = prices_usd[col].astype(float)

# Taux de conversion vers USD
eur_to_usd = 1.08
gbp_to_usd = 1.26
jpy_to_usd = 0.0066
rub_to_usd = 0.011

# Conversion : on remplace les valeurs par leur équivalent en USD
prices_usd['eur'] = prices_usd['eur'] / eur_to_usd
prices_usd['gbp'] = prices_usd['gbp'] / gbp_to_usd
prices_usd['jpy'] = prices_usd['jpy'] * jpy_to_usd
prices_usd['rub'] = prices_usd['rub'] * rub_to_usd

# Ajouter la colonne 'genre principal' du df games à prices_usd

prices_usd = prices_usd.merge(games[['gameid', 'main_genre']], on='gameid', how='left')

# Vérification
prices_usd[['gameid', 'usd', 'eur', 'gbp', 'jpy', 'rub', 'date_acquired', 'plateforme', 'main_genre']].head()
```

### Comparer les jeux les plus achetés par plateformes

```python
# Exposer chaque jeu de la bibliothèque
purchased_exploded = purchased_games.explode('library')

# Extraire la plateforme
purchased_exploded['plateforme'] = purchased_exploded['playerid'].str.split('_').str[-1]

# Renommer la colonne 'library' en 'gameid'
purchased_exploded.rename(columns={'library': 'gameid'}, inplace=True)

# Supprimer les valeurs manquantes
purchased_exploded = purchased_exploded.dropna(subset=['gameid'])

# Ajouter un suffixe pour matcher avec la table 'games'
purchased_exploded['gameid_prefixed'] = purchased_exploded['gameid'].astype(str) + '_' + purchased_exploded['plateforme']

# Compter le nombre total d’achats par jeu et plateforme
nb_achats_plateforme = (
    purchased_exploded
    .groupby(['gameid_prefixed', 'plateforme'])
    .size()
    .reset_index(name='nb_achats')
)

# Nettoyer les espaces et faire le merge avec les métadonnées des jeux
games['gameid'] = games['gameid'].str.strip()
nb_achats_plateforme['gameid_prefixed'] = nb_achats_plateforme['gameid_prefixed'].str.strip()

nb_achats_plateforme = nb_achats_plateforme.merge(
    games[['gameid', 'title', 'main_genre','developers', 'publishers']],
    left_on='gameid_prefixed',
    right_on='gameid',
    how='left'
)

# Top 5 des jeux les plus joués par plateforme
top_jeux_plateforme = (
    nb_achats_plateforme
    .sort_values('nb_achats', ascending=False)
    .groupby('plateforme')
    .head(5)
    .reset_index(drop=True)
)

display(top_jeux_plateforme[['gameid','title', 'main_genre', 'plateforme', 'nb_achats']])
```
