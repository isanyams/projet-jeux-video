#### Nettoyage des datasets games

```python

#Ajout d'une colonne plateforme
games_playstation['plateforme'] = 'playstation'
games_steam['plateforme'] = 'steam'
games_xbox['plateforme'] = 'xbox'

#concatenation des dataset games
games = pd.concat(objs = [games_playstation, games_steam, games_xbox], axis = 0, ignore_index=True)

#remplacement des valeurs vides par unknown
games = games.fillna({'title': 'unknown', 'platform': 'unknown', 'developers' : 'unknown', 'publishers': 'unknown', 'genres': 'unknown', 'supported_languages': 'unknown'})

#Ajout d'une colonne contenant le genre principal
games['main_genre'] = games['genres'].str.strip('[]').str.replace("'", "").str.split(',').str[0]

#Conversion des colonnes publisher, developers et supported_languages en list
games['developers'] = games['developers'].str.strip('[]').str.replace("'", "").str.split(',')
games['publishers'] = games['publishers'].str.strip('[]').str.replace("'", "").str.split(',')
games['supported_languages'] = games['supported_languages'].str.strip('[]').str.replace("'", "").str.split(',')

#nombre de langue par jeu
games['supported_languages_count'] = len(games['supported_languages'])

#conversion de la date release_date en format date
games['release_date'] = pd.to_datetime(arg=games['release_date'])


#ajout d'une colonne année de sortie
games['release_year'] = games['release_date'].dt.year

#l'identifiant du jeu est suffixé par la plateforme
games['gameid'] = games['gameid'].astype('str') + '_' + games['plateforme']
```

#### Nettoyage du dataset price

```python
#Ajout d'une colonne plateforme
prices_playstation['plateforme'] = 'playstation'
prices_steam['plateforme'] = 'steam'
prices_xbox['plateforme'] = 'xbox'

#Concatenation des dataset price
prices = pd.concat([prices_playstation, prices_steam, prices_xbox], axis = 0, ignore_index= True)

#Conversion de la colonne date_acquired en date
prices['date_acquired'] = pd.to_datetime(arg = prices['date_acquired'],errors='coerce')

#Ajout de l'année du prix
prices['acquired_year'] =  prices['date_acquired'].dt.year

#L'identifiant du jeu est suffixé par la plateforme
prices['gameid'] = prices['gameid'].astype('str') + '_' + prices['plateforme']
```

#### Conversion des prix en dollar

```python
# Remplacer les valeurs manquantes par Null

prices = prices.replace(np.nan, 'Null')

# Supprimer les gameid en double dans le dataframe prices

price_without_double = prices.drop_duplicates(subset='gameid')

# Copier le DataFrame original
prices_usd = price_without_double.copy()

# Remplacer 'Null' par np.nan
prices_usd = prices_usd.replace('Null', np.nan)

# Colonnes de devise
cols = ['usd', 'eur', 'gbp', 'jpy', 'rub']

# Convertir les colonnes en float
for col in cols:
    prices_usd[col] = prices_usd[col].astype(float)

# Taux de conversion vers USD
eur_to_usd = 1.08
gbp_to_usd = 1.26
jpy_to_usd = 0.0066
rub_to_usd = 0.011

# Conversion : on remplace les valeurs par leur équivalent en USD
prices_usd['eur'] = prices_usd['eur'] / eur_to_usd
prices_usd['gbp'] = prices_usd['gbp'] / gbp_to_usd
prices_usd['jpy'] = prices_usd['jpy'] * jpy_to_usd
prices_usd['rub'] = prices_usd['rub'] * rub_to_usd

# Ajouter la colonne 'genre principal' du df games à prices_usd

prices_usd = prices_usd.merge(games[['gameid', 'main_genre']], on='gameid', how='left')

# Vérification
prices_usd[['gameid', 'usd', 'eur', 'gbp', 'jpy', 'rub', 'date_acquired', 'plateforme', 'main_genre']].head()
```

## Question sur l'adoption

#### Est-ce que le même jeu est acheté par plusieurs personnes ? (purschased_games)

##### Nombre d'acheteurs par jeu 

```python
# 1. Nettoyage des gameid
all_gameids = purchased_games['library'].dropna().astype(str).str.split(',').explode()
all_gameids = all_gameids.str.strip()                             # Enlève les espaces
all_gameids = all_gameids.str.replace('[^0-9]', '', regex=True)   # Ne garde que les chiffres
all_gameids = all_gameids[all_gameids != ''].astype(str)          # Garde en string (on va ajouter les suffixes après)

# 2. Calculer le nombre d'acheteurs par gameid
gameid_counts = all_gameids.value_counts().reset_index()
gameid_counts.columns = ['gameid', 'nb_acheteurs']

# 3. Ne garder que les jeux achetés plusieurs fois
jeux_achetés_à_plusieurs = gameid_counts[gameid_counts['nb_acheteurs'] > 1].copy()

# 4. Ajouter les gameid avec suffixes pour merge avec 'games'
# On crée une version suffixée pour chaque plateforme si besoin
plateformes = ['playstation', 'xbox', 'steam']

# Liste pour stocker les résultats par plateforme
df_plateformes = []

# Nettoyage de la colonne gameid dans la table games (important pour le merge)
games['gameid'] = games['gameid'].astype(str).str.strip()

for plat in plateformes:
    temp = jeux_achetés_à_plusieurs.copy()
    temp['gameid_suffix'] = temp['gameid'] + f'_{plat}'
    temp['gameid_suffix'] = temp['gameid_suffix'].str.strip()
    temp = temp.merge(
        games[['gameid', 'title']],
        left_on='gameid_suffix',
        right_on='gameid',
        how='left'
    )
    temp['plateforme'] = plat
    df_plateformes.append(temp[['gameid_suffix', 'nb_acheteurs', 'title', 'plateforme']])

# 5. Concaténer tous les résultats
jeux_achetes_plusieurs_fois = pd.concat(df_plateformes, ignore_index=True)

# Résultat
jeux_achetes_plusieurs_fois.head()
```

##### Nombre de fois un même jeu est acheté par playerid

```python
# 1. Exposer la colonne 'library' en une ligne par jeu
purchased_exploded1 = purchased_games.explode('library')

# 2. Ajouter la colonne plateforme en extrayant depuis le playerid
purchased_exploded1['plateforme'] = purchased_exploded1['playerid'].str.split('_').str[-1]

# 3. Renommer la colonne 'library' en 'gameid' pour plus de clarté
purchased_exploded1.rename(columns={'library': 'gameid'}, inplace=True)

# 4. Créer une colonne gameid_prefixed avec suffixe plateforme
purchased_exploded1['gameid_prefixed'] = purchased_exploded1['gameid'].astype(str) + '_' + purchased_exploded1['plateforme']

# 5. Compter combien de fois un même jeu est acheté par playerid
purchased_per_playerid = (
    purchased_exploded1
    .groupby(['playerid', 'gameid_prefixed'])
    .size()
    .reset_index(name='nb_achats')
    .sort_values('nb_achats', ascending=False)
)

# Nettoyer les espaces en trop dans les colonnes de merge
games['gameid'] = games['gameid'].str.strip()
purchased_per_playerid['gameid_prefixed'] = purchased_per_playerid['gameid_prefixed'].str.strip()

# 6. Ajouter les titres des jeux depuis la table 'games'
purchased_per_playerid = purchased_per_playerid.merge(
    games[['gameid', 'title']],
    left_on='gameid_prefixed',
    right_on='gameid',
    how='left'
)

purchased_per_playerid = purchased_per_playerid[['playerid', 'gameid', 'nb_achats', 'title']]
purchased_per_playerid
```

### Nombre de jeux par Rarity ? (PS)

```python
# Calculer la répartition des types de trophées (rarity) pour chaque jeu
rarity_distribution = (
    achievements[achievements['plateforme'] == 'playstation']
    .groupby(['gameid', 'rarity'])['achievementid']
    .count()
    .reset_index(name='count')
)

# Transformer en tableau croisé (un jeu = une ligne, chaque rarity = une colonne)
rarity_pivot = rarity_distribution.pivot(index='gameid', columns='rarity', values='count').fillna(0)

# Calculer la proportion de chaque type par jeu (normalisation ligne par ligne)
rarity_proportion = rarity_pivot.div(rarity_pivot.sum(axis=1), axis=0)

# Fusionner avec les métadonnées des jeux (titre, genre...)
rarity_proportion = rarity_proportion.reset_index().merge(
    games[['gameid', 'title', 'main_genre']], on='gameid', how='left'
)

# Afficher les 10 jeux avec la plus grande proportion de trophées Platinum
top_platinum_games = rarity_proportion.sort_values(by='Platinum', ascending=False)
top_platinum_games
```
