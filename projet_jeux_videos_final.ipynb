#### Nettoyage des datasets games

```python

#Ajout d'une colonne plateforme
games_playstation['plateforme'] = 'playstation'
games_steam['plateforme'] = 'steam'
games_xbox['plateforme'] = 'xbox'

#concatenation des dataset games
games = pd.concat(objs = [games_playstation, games_steam, games_xbox], axis = 0, ignore_index=True)

#remplacement des valeurs vides par unknown
games = games.fillna({'title': 'unknown', 'platform': 'unknown', 'developers' : 'unknown', 'publishers': 'unknown', 'genres': 'unknown', 'supported_languages': 'unknown'})

#Ajout d'une colonne contenant le genre principal
games['main_genre'] = games['genres'].str.strip('[]').str.replace("'", "").str.split(',').str[0]

#Conversion des colonnes publisher, developers et supported_languages en list
games['developers'] = games['developers'].str.strip('[]').str.replace("'", "").str.split(',')
games['publishers'] = games['publishers'].str.strip('[]').str.replace("'", "").str.split(',')
games['supported_languages'] = games['supported_languages'].str.strip('[]').str.replace("'", "").str.split(',')

#nombre de langue par jeu
games['supported_languages_count'] = games['supported_languages'].apply(len)

#reconversion de la colonne supported_languages en string afin d'affecter 0 à la colonne supported_languages_count où la colonne supported_languages vaut unknow
games['supported_languages'] = games['supported_languages'].astype('str')
games.loc[
    games['supported_languages'] == "['unknown']",
    'supported_languages_count'
] = 0

#conversion de la colonne supported_languages en liste
games['supported_languages'] = games['supported_languages'].str.strip('[]').str.replace("'", "").str.split(',')

#conversion de la date release_date en format date
games['release_date'] = pd.to_datetime(arg=games['release_date'])

#ajout d'une colonne année de sortie
games['release_year'] = games['release_date'].dt.year

#l'identifiant du jeu est suffixé par la plateforme
games['gameid'] = games['gameid'].astype('str') + '_' + games['plateforme']
```

#### Nettoyage des dataset players

```python
#Ajout d'une colonne plateforme
players_playstation['plateforme'] = 'playstation'
players_steam['plateforme'] = 'steam'
players_xbox['plateforme'] = 'xbox'

#Concaténation des dataset plateforme
players = pd.concat([players_playstation, players_steam, players_xbox], axis = 0, ignore_index= True)

#Suppression de la colonne nickname
players.drop('nickname', axis=1, inplace=True)

#Conversion de la colonne created au format date
players['created'] = pd.to_datetime(arg = players['created'],errors='coerce')

#Ajout de la colonne année de création
players['created_year'] = players['created'].dt.year

#remplacement des valeurs vide par unknown
players = players.fillna({'country': 'unknown', 'created': 'unknown', 'created_date' : 'unknown', 'created_time' : 'unknown'})

#l'identifiant playerid est suffixé par la plateforme
players['playerid'] = players['playerid'].astype('str') + '_' + players['plateforme']
```

#### Nettoyage du dataset price

```python
#Ajout d'une colonne plateforme
prices_playstation['plateforme'] = 'playstation'
prices_steam['plateforme'] = 'steam'
prices_xbox['plateforme'] = 'xbox'

#Concatenation des dataset price
prices = pd.concat([prices_playstation, prices_steam, prices_xbox], axis = 0, ignore_index= True)

#Conversion de la colonne date_acquired en date
prices['date_acquired'] = pd.to_datetime(arg = prices['date_acquired'],errors='coerce')

#Ajout de l'année du prix
prices['acquired_year'] =  prices['date_acquired'].dt.year

#L'identifiant du jeu est suffixé par la plateforme
prices['gameid'] = prices['gameid'].astype('str') + '_' + prices['plateforme']
```

```python
prices
```

#### Conversion des prix en dollar

```python
# Remplacer les valeurs manquantes par Null

prices = prices.replace(np.nan, 'Null')

# Supprimer les gameid en double dans le dataframe prices

price_without_double = prices.drop_duplicates(subset='gameid')

# Copier le DataFrame original
prices_usd = price_without_double.copy()

# Remplacer 'Null' par np.nan
prices_usd = prices_usd.replace('Null', np.nan)

# Colonnes de devise
cols = ['usd', 'eur', 'gbp', 'jpy', 'rub']

# Convertir les colonnes en float
for col in cols:
    prices_usd[col] = prices_usd[col].astype(float)

# Taux de conversion vers USD
eur_to_usd = 1.08
gbp_to_usd = 1.26
jpy_to_usd = 0.0066
rub_to_usd = 0.011

# Conversion : on remplace les valeurs par leur équivalent en USD
prices_usd['eur'] = prices_usd['eur'] / eur_to_usd
prices_usd['gbp'] = prices_usd['gbp'] / gbp_to_usd
prices_usd['jpy'] = prices_usd['jpy'] * jpy_to_usd
prices_usd['rub'] = prices_usd['rub'] * rub_to_usd

# Ajouter la colonne 'genre principal' du df games à prices_usd

prices_usd = prices_usd.merge(games[['gameid', 'main_genre']], on='gameid', how='left')

# Vérification
prices_usd[['gameid', 'usd', 'eur', 'gbp', 'jpy', 'rub', 'date_acquired', 'plateforme', 'main_genre']].head()
```

#### Nettoyage du dataset review

```python
#Remplissage des valeurs vide par unknown
reviews = reviews.fillna({'review': 'unknown'})

#Ajout de _steam aux identifiants gameid et playerid
reviews['gameid'] = reviews['gameid'].astype('str') + '_steam'
reviews['playerid'] = reviews['playerid'].astype('str') + '_steam'

# Crée une nouvelle colonne avec le nombre d'avis par jeu dans le dataframe reviews
reviews['reviews_count'] = reviews['gameid'].map(reviews['gameid'].value_counts())

reviews
```

#### Succès économique

##### Nombre d’exemplaires de jeux vendus par année ? par plateforme ? par genre ? Quelle est la corrélation entre le prix et le nombre de ventes ?

```python
#Explode pour récupérer chaque jeu qu'un joueur a acheté
purchased_games_explode_library = purchased_games.explode('library')
purchased_games_explode_library.rename(columns=
    {
        'library':'gameid'
    },
    inplace=True
)

#ajout du suffixe de la plateforme dans l'identifiant du jeu
purchased_games_explode_library['gameid'] = purchased_games_explode_library['gameid'].astype('str') + '_' + purchased_games_explode_library['plateforme']
purchased_games_explode_library
```

```python
#jointure des prix, des jeux et des jeux achetés
purchased_games_price_join = pd.merge(left=prices_usd[['gameid','usd','date_acquired']], right=purchased_games_explode_library, on='gameid', how='inner')
purchased_games_price_join = pd.merge(left = games[['gameid','title','main_genre']], right = purchased_games_price_join, on ='gameid', how='inner')

purchased_games_price_join['year_acquired'] = purchased_games_price_join['date_acquired'].dt.year
purchased_games_price_join
```

```python
#nombre de ventes par plateforme
nb_sale_by_plateforme = purchased_games_price_join.groupby(['plateforme']).size().reset_index(name='nb_sale').sort_values(by='nb_sale', ascending=False)
nb_sale_by_plateforme
```

```python
nb_sale_by_plateforme.to_csv('out/nb_sale_by_plateforme.csv')
```

```python
#nombre de ventes par genre
nb_sale_by_main_genre = purchased_games_price_join.groupby(['main_genre']).size().reset_index(name='nb_sale').sort_values(by='nb_sale', ascending=False)
nb_sale_by_main_genre.head(50)
```

```python
nb_sale_by_main_genre.to_csv('out/nb_sale_by_main_genre.csv')
```

```python
#Nombre de vente par année
nb_sale_by_year = purchased_games_price_join.groupby(['year_acquired']).size().reset_index(name='nb_sale').sort_values(by='nb_sale', ascending=False)
nb_sale_by_year
```

```python
nb_sale_by_year.to_csv('out/nb_sale_by_year.csv')
```

```python
#correlation entre le nombre de vente et le prix
sales_and_prices = purchased_games_price_join.groupby(['gameid','usd']).size().reset_index(name='nb_sale').sort_values(by='nb_sale', ascending=False)
sales_and_prices.head(50)
```

```python
sales_and_prices.to_csv('out/sales_and_prices.csv')
```

```python
#Nombre d'achievement par année, plateforme, joueur et jeu
achievement_year_player_game_plateforme = pd.read_csv('achievement_year_player_game_plateforme.csv')
achievement_year_player_game_plateforme.rename(columns=
    {
        'PLAYERID':'playerid',
        'GAMEID':'gameid',
        'ACQUIRED_YEAR':'acquired_year',
        'TITLE':'title',
        'PLATEFORME':'plateforme',
        'NB_ACHIVEMENT':'nb_achievement'
    },
    inplace=True
)
achievement_year_player_game_plateforme.sort_values(by='nb_achievement', ascending=False).head(50)
```

##### Nombre total d'achievement possible par jeu

```python
achievements_by_game = achievements.groupby(['gameid']).size().reset_index(name = 'nb_achievement_total_by_game')
achievements_by_game.head()
```

##### Taux d'achivement par année, jeu, joueur et plateforme

```python
achievement_year_player_game_plateforme = pd.merge(left=achievement_year_player_game_plateforme, right=achievements_by_game, on='gameid', how='left')
achievement_year_player_game_plateforme['taux_achievement'] = achievement_year_player_game_plateforme['nb_achievement'] / achievement_year_player_game_plateforme['nb_achievement_total_by_game']


achievement_year_player_game_plateforme.sort_values(by='taux_achievement', ascending=False).head(50)
```

```python
achievement_year_player_game_plateforme.to_csv('out/achievement_year_player_game_plateforme.csv')
```

##### Taux d'achievement par jeu

```python
nb_achivement_by_game = achievement_year_player_game_plateforme.groupby('gameid')['nb_achievement'].sum().reset_index(name='nb_achivement_by_game')
nb_achivement_by_game = pd.merge(left=nb_achivement_by_game, right=achievements_by_game, on='gameid', how='left')
nb_achivement_by_game['taux_achievement'] = nb_achivement_by_game['nb_achivement_by_game'] / nb_achivement_by_game['nb_achievement_total_by_game']
nb_achivement_by_game['plateforme'] = nb_achivement_by_game['gameid'].str.split('_').str[1]

nb_achivement_by_game.sort_values(by='taux_achievement', ascending=False).head(50)
```

```python
nb_achivement_by_game.to_csv('out/nb_achivement_by_game.csv')
```
