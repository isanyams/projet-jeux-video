#### Nettoyage des datasets games

```python

#Ajout d'une colonne plateforme
games_playstation['plateforme'] = 'playstation'
games_steam['plateforme'] = 'steam'
games_xbox['plateforme'] = 'xbox'

#concatenation des dataset games
games = pd.concat(objs = [games_playstation, games_steam, games_xbox], axis = 0, ignore_index=True)

#remplacement des valeurs vides par unknown
games = games.fillna({'title': 'unknown', 'platform': 'unknown', 'developers' : 'unknown', 'publishers': 'unknown', 'genres': 'unknown', 'supported_languages': 'unknown'})

#Ajout d'une colonne contenant le genre principal
games['main_genre'] = games['genres'].str.strip('[]').str.replace("'", "").str.split(',').str[0]

#Conversion des colonnes publisher, developers et supported_languages en list
games['developers'] = games['developers'].str.strip('[]').str.replace("'", "").str.split(',')
games['publishers'] = games['publishers'].str.strip('[]').str.replace("'", "").str.split(',')
games['supported_languages'] = games['supported_languages'].str.strip('[]').str.replace("'", "").str.split(',')

#nombre de langue par jeu
games['supported_languages_count'] = games['supported_languages'].apply(len)

#reconversion de la colonne supported_languages en string afin d'affecter 0 à la colonne supported_languages_count où la colonne supported_languages vaut unknow
games['supported_languages'] = games['supported_languages'].astype('str')
games.loc[
    games['supported_languages'] == "['unknown']",
    'supported_languages_count'
] = 0

#conversion de la colonne supported_languages en liste
games['supported_languages'] = games['supported_languages'].str.strip('[]').str.replace("'", "").str.split(',')

#conversion de la date release_date en format date
games['release_date'] = pd.to_datetime(arg=games['release_date'])

#ajout d'une colonne année de sortie
games['release_year'] = games['release_date'].dt.year

#l'identifiant du jeu est suffixé par la plateforme
games['gameid'] = games['gameid'].astype('str') + '_' + games['plateforme']
```

#### Nettoyage du dataset price

```python
#Ajout d'une colonne plateforme
prices_playstation['plateforme'] = 'playstation'
prices_steam['plateforme'] = 'steam'
prices_xbox['plateforme'] = 'xbox'

#Concatenation des dataset price
prices = pd.concat([prices_playstation, prices_steam, prices_xbox], axis = 0, ignore_index= True)

#Conversion de la colonne date_acquired en date
prices['date_acquired'] = pd.to_datetime(arg = prices['date_acquired'],errors='coerce')

#Ajout de l'année du prix
prices['acquired_year'] =  prices['date_acquired'].dt.year

#L'identifiant du jeu est suffixé par la plateforme
prices['gameid'] = prices['gameid'].astype('str') + '_' + prices['plateforme']
```

```python
prices
```

#### Conversion des prix en dollar

```python
# Remplacer les valeurs manquantes par Null

prices = prices.replace(np.nan, 'Null')

# Supprimer les gameid en double dans le dataframe prices

price_without_double = prices.drop_duplicates(subset='gameid')

# Copier le DataFrame original
prices_usd = price_without_double.copy()

# Remplacer 'Null' par np.nan
prices_usd = prices_usd.replace('Null', np.nan)

# Colonnes de devise
cols = ['usd', 'eur', 'gbp', 'jpy', 'rub']

# Convertir les colonnes en float
for col in cols:
    prices_usd[col] = prices_usd[col].astype(float)

# Taux de conversion vers USD
eur_to_usd = 1.08
gbp_to_usd = 1.26
jpy_to_usd = 0.0066
rub_to_usd = 0.011

# Conversion : on remplace les valeurs par leur équivalent en USD
prices_usd['eur'] = prices_usd['eur'] / eur_to_usd
prices_usd['gbp'] = prices_usd['gbp'] / gbp_to_usd
prices_usd['jpy'] = prices_usd['jpy'] * jpy_to_usd
prices_usd['rub'] = prices_usd['rub'] * rub_to_usd

# Ajouter la colonne 'genre principal' du df games à prices_usd

prices_usd = prices_usd.merge(games[['gameid', 'main_genre']], on='gameid', how='left')

# Vérification
prices_usd[['gameid', 'usd', 'eur', 'gbp', 'jpy', 'rub', 'date_acquired', 'plateforme', 'main_genre']].head()
```

#### Review

#### Nombre d’avis par jeu ? par genre ? par pays ? par joueur ? (steam) 

```python
#jointure des dataset reviews, games et players
games_review_join = pd.merge(reviews[['reviewid','playerid','gameid']], games[['gameid', 'title', 'main_genre']], on='gameid', how='left')
players_games_review_join = pd.merge(games_review_join, players[['playerid','country']], on='playerid', how='left')

players_games_review_join.head(50)

```

```python
#Nombre de revue par genre
review_by_genre= players_games_review_join.groupby(['main_genre'])['reviewid'].size().reset_index(name='nb_review').sort_values(by='nb_review', ascending=False)

#taux de revue par pays
games_genre = games.copy()

review_by_genre['nb_game_by_genre'] = review_by_genre.apply(
    lambda x : games_genre[(games_genre['main_genre'] == x['main_genre']) & (games_genre['plateforme'] == 'steam')].shape[0],
    axis=1
)

review_by_genre['rate_review_by_genre'] = review_by_genre['nb_review'] / review_by_genre['nb_game_by_genre']

review_by_genre.sort_values(by='rate_review_by_genre', ascending=False)
```

```python
#Nombre de jeu unique steam acheté
nb_purchased_games_steam_by_gameid = purchased_games_explode_library[purchased_games_explode_library['plateforme'] == 'steam'].groupby('gameid')['games_bought'].sum().reset_index()
nb_purchased_games_steam_by_gameid.head()
```

```python
#Nombre de revue par jeu
review_by_jeu = players_games_review_join.groupby(['gameid','title','main_genre'])['reviewid'].size().reset_index(name='nb_review').sort_values(by='nb_review', ascending=False)
review_by_jeu = pd.merge(left=review_by_jeu, right=nb_purchased_games_steam_by_gameid, on='gameid', how='left')
review_by_jeu['review_by_game_rate'] = review_by_jeu['nb_review'] / review_by_jeu['games_bought']

#review_by_jeu = review_by_jeu.replace(np.nan, 'Null')
review_by_jeu.head(50).sort_values(by='review_by_game_rate', ascending=False)
```

```python
#Nombre de revue par pays
review_by_country = players_games_review_join.groupby(['country'])['reviewid'].size().reset_index(name='nb_review').sort_values(by='nb_review', ascending=False)

#suppression des pays non connus
review_by_country = review_by_country[review_by_country['country'] != 'unknown']

players_country = players.copy()

#nombre de joueur par pays
review_by_country['nb_player_by_country'] = review_by_country.apply(
    lambda x : players_country[(players_country['country'] == x['country']) & (players_country['plateforme'] == 'steam')].shape[0],
    axis=1
)

#taux de revue par pays
review_by_country['rate_review_by_country'] = review_by_country['nb_review'] / review_by_country['nb_player_by_country']

review_by_country.sort_values(by='rate_review_by_country', ascending=False).head(50)
```

```python
#Nombre de jeu steam acheté par joueur
nb_purchased_games_steam_by_playerid = purchased_games_explode_library[purchased_games_explode_library['plateforme'] == 'steam'].groupby('playerid')['games_bought'].sum().reset_index()
nb_purchased_games_steam_by_playerid.head()
```

```python
#Nombre de revues par joueur
review_by_player = players_games_review_join.groupby(['playerid'])['reviewid'].size().reset_index(name='nb_review').sort_values(by='nb_review', ascending=False)
review_by_player = pd.merge(left=review_by_player, right=nb_purchased_games_steam_by_playerid, on='playerid', how='left')
review_by_player['review_by_player_rate'] = review_by_player['nb_review'] / review_by_player['games_bought']

#review_by_player = review_by_player.replace(np.nan, 'Null')
review_by_player.head(50).sort_values(by='review_by_player_rate', ascending=False)

review_by_player[review_by_player['games_bought'].notnull()].head(50)
```
