## Réponse à la problématique

#### 1. Facteur économique : Revenu total d’un jeu

```python
# Fusion prix et jeux achetés
revenu_jeu = prices_usd[['gameid', 'usd']].copy()

# Verif que chaque colonne a le même type
purchased_games['library'] = purchased_games['library'].astype(str)
revenu_jeu['gameid'] = revenu_jeu['gameid'].astype(str)

# Explode la colonne library et merge
revenu_jeu = revenu_jeu.merge(purchased_games.explode('library'), left_on='gameid', right_on='library', how='left')

# Group by gameid et calculer le total revenue
revenu_jeu = revenu_jeu.groupby('gameid')['usd'].sum().reset_index(name='revenue_usd')
revenu_jeu = revenu_jeu.sort_values(by='revenue_usd', ascending=False)

# Affichage des 10 jeux les plus rentables
revenu_jeu.head(10)
```

```python
# Convertir les colonnes 'gameid' en chaînes de caractères pour éviter les erreurs de type
revenu_jeu['gameid'] = revenu_jeu['gameid'].astype(str)
games['gameid'] = games['gameid'].astype(str)

# Ajouter le nom du jeu à la table
revenu_jeu = revenu_jeu.merge(games[['gameid', 'title']], on='gameid', how='left')

# Réorganiser les colonnes pour une meilleure lisibilité
revenu_jeu = revenu_jeu[['gameid', 'title', 'revenue_usd']]

# Afficher les 10 jeux les plus rentables
revenu_jeu.head(10)
```

### 2. Adoption / popularité : Nombre d’acheteurs par jeu

```python
# 1. Créer une copie du dataframe avec uniquement les colonnes nécessaires
popularity = purchased_games[['playerid', 'library']].copy()

# 2. S'assurer que la colonne 'library' contient bien des listes
popularity['library'] = popularity['library'].apply(lambda x: eval(x) if isinstance(x, str) and x != 'nan' else np.nan)

# 3. Exploser la colonne pour que chaque ligne corresponde à un jeu
popularity_exploded = popularity.explode('library')

# 4. Convertir les gameid en string pour uniformiser (si nécessaire)
popularity_exploded['library'] = popularity_exploded['library'].astype(str)

# 5. Compter le nombre de joueurs ayant chaque jeu
popularity_count = popularity_exploded.groupby('library').size().reset_index(name='player_count')

# 6. Renommer proprement
popularity_count = popularity_count.rename(columns={'library': 'gameid'})

# Trier les jeux les plus populaires
popularity_count = popularity_count.sort_values('player_count', ascending=False)
popularity_count.head(10)
```

### Engagement : temps moyen de jeu par joueur 

```python
# 1. Extraire les colonnes playerid et gameid depuis achievements et history
achievement_game_map = achievements[['achievementid', 'gameid']].drop_duplicates()
history_with_gameid = history.merge(achievement_game_map, on='achievementid', how='left')

# 2. Calculer le temps total joué par joueur
history['time_seconds'] = history['time'].apply(lambda x: x.hour * 3600 + x.minute * 60 + x.second if pd.notnull(x) else 0)
total_time = history.groupby('playerid')['time_seconds'].sum().reset_index()
total_time['hours_played'] = total_time['time_seconds'] / 3600

# 3. Calculer le nombre de jours joués par joueur
total_time['days_played'] = history.groupby('playerid')['date'].nunique().values

# 4. Calculer le temps moyen de jeu par jour
total_time['avg_hours_per_day'] = total_time['hours_played'] / total_time['days_played']
total_time = total_time[total_time['avg_hours_per_day'].between(0, 24)]

# 5. Associer playerid à gameid
player_game = history_with_gameid[['playerid', 'gameid']].drop_duplicates()
player_game_playtime = player_game.merge(total_time[['playerid', 'avg_hours_per_day']], on='playerid', how='inner')

# 6. Moyenne du temps moyen par jeu
avg_playtime_per_game = player_game_playtime.groupby('gameid')['avg_hours_per_day'].mean().reset_index()
avg_playtime_per_game.rename(columns={'avg_hours_per_day': 'avg_playtime_per_player'}, inplace=True)

# 7. Voir les jeux les plus engageants
avg_playtime_per_game.sort_values('avg_playtime_per_player', ascending=False).head(10)
```

### 4. Réputation : Nombre moyen d’avis positifs par jeu (biaisé)

```python
# Calculer le nombre d'avis par jeu

reputation = reviews.groupby('gameid')['reviews_count'].sum().reset_index()
reputation = reputation.rename(columns={'reviews_count': 'total_reviews'})

# Convertir les colonnes 'gameid' en chaînes de caractères pour éviter les erreurs de type
reputation['gameid'] = reputation['gameid'].astype(str)
games['gameid'] = games['gameid'].astype(str)

# Ajouter le titre du jeu grâce au gameid et à la table games
reputation = pd.merge(reputation, games[['gameid', 'title']], on='gameid')

reputation.head()
```
