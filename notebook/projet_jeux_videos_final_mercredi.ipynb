Purchased_games

```python
# ajouter une colonne plateforme pour playstation

df_purchased_games_playstation["plateforme"]="playstation"
df_purchased_games_playstation.head()

# ajouter une colonne plateforme pour steam

df_purchased_games_steam["plateforme"]="steam"
df_purchased_games_steam.head()

# ajouter une colonne plateforme pour xbox

df_purchased_games_xbox["plateforme"]="xbox"
df_purchased_games_xbox.head()

# concatenation et avec ignore rajouter index en suivant bon ordre

purchased_games=pd.concat([df_purchased_games_playstation,df_purchased_games_steam,df_purchased_games_xbox], ignore_index=True)

purchased_games.shape

# ajouter une colonne count
purchased_games['games_bought'] = purchased_games['library'].str.split(',').str.len()

purchased_games
```

```python
# regarder type de premiere ligne de la colonne library

type(purchased_games['library'].iloc[0])

# calculer le nombre des valeurs manquantes 

purchased_games.isnull().sum()

# mettre les valeurs manquantes en liste pour garder les joueurs sans jeu pour sortir le nombre de joueurs actif par plateforme


purchased_games["library"]=purchased_games["library"].apply (
                                        lambda x:                                         
                                          "[]" if pd.isnull(x)
                                          else x                                       
                                          
)

purchased_games.head()

# vérifier le nombre des valeurs manquantes 

purchased_games.isnull().sum()


# mettre tout en type liste pour pouvoir faire explode après

purchased_games['library'] = purchased_games['library'].apply(ast.literal_eval)
```

```python
# split (avec explode) pour transformer la liste des valeurs en autant de lignes qu'il y a d'élément dans la liste 
# pour deux buts: 
# 1) nb de jeux par joueur (avec groupby et sum)
# 2) nb de joueurs par jeu (avec groupby et sum)

purchased_games_splitlibrary = purchased_games.explode('library')

purchased_games_splitlibrary

# le nombre de joueurs n'ayant aucun jeu

purchased_games_splitlibrary.isnull().sum()

# renommer la colonne library en gameid

purchased_games_splitlibrary.columns =["playerid", "gameid", "plateforme"]

purchased_games_splitlibrary.head()
```

Players

```python
# Players

df_players_playstation['platforme'] = 'playstation'
df_players_steam['platforme'] = 'steam'
df_players_xbox['platforme'] = 'xbox'
df_players = pd.concat([df_players_playstation, df_players_steam, df_players_xbox], axis = 0, ignore_index= True)
df_players.drop('nickname', axis=1, inplace=True)
df_players['created'] = pd.to_datetime(arg = df_players['created'],errors='coerce')
df_players['created_date'] = df_players['created'].dt.strftime('%Y-%m-%d')
df_players['created_time'] = df_players['created'].dt.strftime('%H:%M:%S')
df_players = df_players.fillna({'country': 'unknown', 'created': 'unknown', 'created_date' : 'unknown', 'created_time' : 'unknown'})

df_players['playerid'] = df_players.apply(
    lambda x: str(x['playerid'])+''+x['platforme'],
    axis= 1
)
```

Prices

```python
df_prices_playstation['platforme'] = 'playstation'
df_prices_steam['platforme'] = 'steam'
df_prices_xbox['platforme'] = 'xbox'
df_prices = pd.concat([df_prices_playstation, df_prices_steam, df_prices_xbox], axis = 0, ignore_index= True)

df_prices = df_prices.fillna({'usd': -1, 'eur': -1, 'gbp' : -1, 'jpy': -1, 'rub': -1})
df_prices['date_acquired'] = pd.to_datetime(arg = df_prices['date_acquired'],errors='coerce')
df_prices['date_acquired'] = pd.to_datetime(arg = df_prices['date_acquired'],errors='coerce')
df_prices['gameid'] = df_prices.apply(
    lambda x: str(x['gameid'])+''+x['platforme'],
    axis= 1
)

df_prices['year_acquired'] = df_prices['date_acquired'].apply(
    lambda x: x.year
)

df_prices.head()
```
