```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
play_history=pd.read_csv("src/game/playstation/history.csv")
play_history.head()
```

```python
pd.set_option("display.max_columns", None)
steam_history=pd.read_csv("src/game/steam/history.csv")
steam_history.head()
```

```python
pd.set_option("display.max_columns", None)
xbox_history=pd.read_csv("src/game/xbox/history.csv")
xbox_history.head()
```

```python
# ajouter une colonne plateforme pour playstation

play_history["plateforme"]="playstation"
play_history.head()
```

```python
# ajouter une colonne plateforme pour steam

steam_history["plateforme"]="steam"
steam_history.head()
```

```python
# ajouter une colonne plateforme pour xbox

xbox_history["plateforme"]="xbox"
xbox_history.head()
```

```python
# vérifier le type de la colonne date_acquired

play_history.info()
```

```python
steam_history.info()
```

```python
xbox_history.info()
```

```python
# transformer string en datetime de la colonne date_acquired (playstation)

play_history["date_acquired"]=pd.to_datetime(play_history['date_acquired'])
```

```python
play_history.info()
```

```python
# transformer string en datetime de la colonne date_acquired (steam)

steam_history["date_acquired"]=pd.to_datetime(steam_history["date_acquired"])
steam_history.info()
```

```python
# transformer string en datetime de la colonne date_acquired (xbox)

xbox_history["date_acquired"]=pd.to_datetime(xbox_history["date_acquired"])
xbox_history.info()
```

```python
# concatenation et avec ignore rajouter index en suivant bon ordre

history=pd.concat([play_history,steam_history,xbox_history], ignore_index=True)
```

```python
history
```

```python
# spliter date_acquired en date et heure

history['date'] =history['date_acquired'].dt.date
history['heure'] = history['date_acquired'].dt.time

history
```

```python
history.columns
```

```python
# changer l'ordre des colonnes

history=history[['playerid', 'achievementid', 'date_acquired', 'date', 'heure', 'plateforme']]
```

```python
history.head()
```

```python
# fichier final

history.to_csv("history.csv")
```
