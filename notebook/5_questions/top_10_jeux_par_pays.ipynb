```python
import pandas as pd
import ast
```

### Top 10 des jeux par pays ?

```python
games=pd.read_csv("final/games.csv")
```

```python
players=pd.read_csv("final/players.csv")
```

```python
purchased_games=pd.read_csv("final/purchased_games.csv", index_col=0)
```

```python
games.head()
```

```python
purchased_games.head()
```

```python
# nettoyer les lignes qui n'ont pas de valeur pour creer un string qui simule une liste avec une valeur par defaut unknown 
purchased_games["library"]=purchased_games["library"].apply (
                                        lambda x:                                         
                                          "['unknown']" if pd.isnull(x)
                                          else x                                       
                                          
)

purchased_games.head()
```

```python
# mettre tout en type list pour pouvoir faire explode après

purchased_games['library'] = purchased_games['library'].apply(ast.literal_eval)
```

```python
# explode de la ligne library

purchased_games_splitlibrary = purchased_games.explode('library')
```

```python
purchased_games_splitlibrary.head()
```

```python
# renommer la colonne library en gameid

purchased_games_splitlibrary.columns =["playerid", "gameid", "plateforme"]

purchased_games_splitlibrary.head()
```

```python
#mettre identifiant du jeu suffixé par le plateforme pour faire merge
purchased_games_splitlibrary['gameid'] = purchased_games_splitlibrary['gameid'] + '_' + purchased_games_splitlibrary['plateforme']
```

```python
purchased_games_splitlibrary.head()
```

```python
# enregistrer en csv pour la réutiliser dans d' autres questions
purchased_games_splitlibrary.to_csv("powerbi/purchased_games_splitlibrary.csv")
```

```python
# reduire le nombre de colonnes de df games pour garder que le gameid, title et plateforme afin de s'assurer qu'il y 
# a assez de memoire pour merge

games=games[['gameid', 'title','plateforme']]
games.head()
```

```python
# merge avec purchased game splité sur le gameid, left c'est games 

games_purchased_games=games.merge(
  purchased_games_splitlibrary,
  left_on =['gameid'],
  right_on =['gameid'],
  how='left'    
)
```

```python
players.head()
```

```python
# reduire le nombre de colonnes 

players=players[['playerid', 'country']]
players.head()
```

```python
# seconde merge avec players sur playerid 
# reduire le nombre de colonnes

games_purchasedgames_players=games_purchased_games.merge(
  players,
  left_on =['playerid'],
  right_on =['playerid'],
  how='left'    
)
```

```python
games_purchasedgames_players
```

```python
# vérifier les valeur manquantes
games_purchasedgames_players.isnull().sum()
```

```python
# verifier les nombre des jeux sans joueurs sur les différentes plateformes

mask = games_purchasedgames_players["plateforme_x"] == "playstation"
masked_df=games_purchasedgames_players[mask]
masked_df.isnull().sum()
```

```python
mask = games_purchasedgames_players["plateforme_x"] == "steam"
masked_df=games_purchasedgames_players[mask]
masked_df.isnull().sum()
```

```python
mask = games_purchasedgames_players["plateforme_x"] == "xbox"
masked_df=games_purchasedgames_players[mask]
masked_df.isnull().sum()
```

```python
# Requete finale pour répondre a la quesiton : Top 10 des jeux par pays ?

top_games = (games_purchasedgames_players
    .groupby(['gameid', 'title', 'country','plateforme_x'])  
    .size()                                  
    .reset_index(name='count')              
    .sort_values(['country', 'count'], ascending=[True, False])  
    .groupby('country')                     
    .head(10)                              
)
```

```python
top_games
```

```python
# enregistrer en csv 

top_games.to_csv("powerbi/top_10_jeux_par_pays.csv")
```
