```python
import pandas as pd
```

### Ã‰volution des prix dans le temps par genre ? 

```python
games=pd.read_csv("final/games.csv")
games.head()
```

```python
prices=pd.read_csv("final/prices.csv")
prices.head()
```

```python
# merger deux tables 

games_prices=games.merge(
  prices,
  left_on =['gameid'],
  right_on =['gameid'],
  how='left'    
)
```

```python
games_prices.head()
```

```python
# creer une table avec les colonnes dont on a besoin pour rÃ©pondre a la quesiton 

games_prices=games_prices[['gameid', 'main_genre', 'usd','eur','gbp','jpy','rub','date_acquired','release_date']]

games_prices.head()
```

```python
# converir release_date en datetime et extraire l'annÃ©e

games_prices['year'] = pd.to_datetime(games_prices['release_date']).dt.year
```

```python
games_prices.head()
```

```python
# Calculer la moyenne des prix par annÃ©e et par genre. RÃ©ponse Ã  la question : 
# Ã‰volution des prix dans le temps par genre ? 

avg_price_by_year_genre = (games_prices
    .groupby(['year', 'main_genre']) 
    [['usd','eur','gbp','jpy','rub']]
    .mean()                        
    .reset_index()                 
    .round(2)                      
)

avg_price_by_year_genre
```

```python
avg_price_by_year_genre.to_csv("powerbi/prix_dans_temps_par_genre.csv")
```
