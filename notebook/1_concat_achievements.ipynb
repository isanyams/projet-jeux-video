```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
play_achievement=pd.read_csv("src/game/playstation/achievements.csv")
play_achievement.head()
```

```python
pd.set_option("display.max_columns",None)
steam_achievements=pd.read_csv("src/game/steam/achievements.csv")
steam_achievements.head()
```

```python
pd.set_option("display.max_columns",None)
xbox_achievements=pd.read_csv("src/game/xbox/achievements.csv")
xbox_achievements.head()
```

```python
# ajouter les colonnes pour playstation

play_achievement["points"]=None
play_achievement["plateforme"]="playstation"

play_achievement.head()
```

```python
# ajouter les colonnes pour steam

steam_achievements["rarity"]="unknown"
steam_achievements["points"]=None
steam_achievements["plateforme"]="steam"

steam_achievements.head()
```

```python
# ajouter les colonnes pour xbox

xbox_achievements["rarity"]="unknown"
xbox_achievements["plateforme"]="xbox"

xbox_achievements.head()
```

```python
# changer l'ordre des colonnes xbox

xbox_achievements=xbox_achievements[['achievementid', 'gameid', 'title', 'description', 'rarity', 'points', 'plateforme']]
xbox_achievements.head()
```

```python
# concatenation et avec ignore rajouter index en suivant bon ordre

achievements=pd.concat ([play_achievement, steam_achievements, xbox_achievements], ignore_index= True )
```

```python
achievements
```

```python
# verifier les valeurs manquantes

achievements.isnull().sum()
```

```python
# test: remplacer valeur null par unknown en ajoutant la colonne title_1

# achievements["title_1"] = achievements ["title"].apply (
#                                   lambda x: 
#                                     "unknown" if pd.isnull(x)
#                                     else x
#                                   )                                    
```

```python
# remplacer valeur null par unknown dans la colonne title

achievements["title"] = achievements ["title"].apply (
                                  lambda x: 
                                    "unknown" if pd.isnull(x)
                                    else x
                                  )   
```

```python
# remplacer valeur null par unknown dans la colonne decription

achievements["description"] = achievements ["description"].apply (
                                  lambda x: 
                                    "unknown" if pd.isnull(x)
                                    else x
                                  ) 

```

```python
# vérification des valeurs manquantes. c'est normal qu'il y en a dans la colonne points car nous avons decidé de mettre Nan là,
# où il n'y a pas de valeurs numériques

achievements.isnull().any()
```

```python
# fichier final

achievements.to_csv("achievements.csv")
```
